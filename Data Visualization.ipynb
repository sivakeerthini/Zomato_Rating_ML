```python
carlist = []
for car in subdf['brand_name'].unique(): 
    sublist = []
    petrolavg = subdf[(subdf['brand_name'] == car) & (subdf['fuel'] == 'Petrol')]['selling_price'].mean()
    dieselavg = subdf[(subdf['brand_name'] == car) & (subdf['fuel'] == 'Diesel')]['selling_price'].mean()
    sublist.extend([car, petrolavg, dieselavg])
    carlist.append(sublist)
```
